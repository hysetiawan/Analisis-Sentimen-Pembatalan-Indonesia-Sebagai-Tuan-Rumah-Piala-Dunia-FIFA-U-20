```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

```python
from textblob import TextBlob
import re
def clean_tweet(tweet):
  return ' '.join(re.sub("(@[A-Za-z0-9]+) | (@[^0-9A-Za-z \t]) | (/w+:\/\/\S+)", " ", tweet).split())
#Sentimen analsis dengan polarity menggunakan library textblob
def analize_sentiment(tweet):
  analysis = TextBlob (clean_tweet(tweet))
  if analysis.sentiment.polarity > 0:
    return 1
  elif analysis.sentiment.polarity == 0:
    return 0
  else :
    return -1
  def analize_subjectorobject(tweet):
    #Kita dapat melihat polaritas [-1][0][1] untuk -1 termasuk negatif, 0 netral dan 1 positif.
    #Misal terdapat polarity 0.8 yang berarti pernyataan positif dan subjektifitas 0.75 merujuk bahwa itu adalah opini publik dan bukan informasi faktual
    analysis = TextBlob(clean_tweet(tweet))
    if analysis.sentiment.polarity > 0:
      return "Positif"
    elif analysis.sentiment.polarity == 0:
      return "Netral"
    else:
      return "Negatif"
```

```python
# buat daftar untuk menambah data dari twitter
tweets_list = []
```

```python
#buat TwiterSearcScrapper untuk mengambil data
for i, tweet in enumerate(sntwitter.TwitterSearchScraper('fifa batal u-20 since:2023-03-25 until:2023-03-30').get_items()):
  if i>20000:
    break
  tweets_list.append([tweet.date, tweet.id, tweet.content, tweet.user.username, analize_sentiment(tweet.content)])
```

```python
# membuat tabel dari data tweet yang sudah di ambil
tweet = pd.DataFrame(tweets_list, columns=['Datetime', 'Tweet Id', 'Tweet', 'Username', 'Label'])
```

```python
tweet_positif=tweet[tweet['Label'] > 0]
tweet_positif.head()
```

```python
tweet_negatif=tweet[tweet['Label'] < 0]
tweet_negatif.head()
```

```python
tweet_netral=tweet[tweet['Label'] == 0]
tweet_netral.head()
```

```python
frames = [tweet_positif, tweet_negatif, tweet_netral]
#menghapus duplikasi
hasil = pd.concat(frames).drop_duplicates().reset_index(drop=True)
```
